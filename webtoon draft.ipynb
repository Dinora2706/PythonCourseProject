```python
grp = data.groupby("Genre")
x = grp["id"].agg(np.mean)
y = grp["Subscribers"].agg(np.sum)
z = grp["Likes"].agg(np.sum)
print(x)
print(y)
print(z)
```

```python
plt.figure(figsize = (12, 5))
plt.plot(x, "ro")
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (16, 5))
plt.plot(x, "ro", color = "b")
plt.xticks(rotation = 90)
plt.title("Genre vs Likes")
plt.xlabel("Genre")
plt.ylabel("Likes")
plt.show()
```
